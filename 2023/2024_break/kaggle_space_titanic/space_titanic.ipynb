```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
```

```python
y = train_data["Transported"]
features = ["HomePlanet", "CryoSleep", "Destination", "Age", "VIP"]

X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
# num_features = ["Age"]
# cat_features = ["HomePlanet", "CryoSleep", "Destination", "VIP"]

# # Fit the imputer on the training data
# num_imputer = SimpleImputer(strategy="mean")
# cat_imputer = SimpleImputer(strategy="mean")

# my_imputer.fit(train_data[features])

# # Transform the training and test data
# imp_train = pd.DataFrame(my_imputer.transform(train_data[features]))
# imp_test = pd.DataFrame(my_imputer.transform(test_data[features]))

# # Assign column names
# imp_train.columns = train_data[features].columns
# imp_test.columns = test_data[features].columns
```

```python
import xgboost as xgb
from xgboost import DMatrix
```

```python
# Prepare data in XGBoost
xg_data = DMatrix(X, label = y)

params = {
    "colsample_bynode": 0.8,  # Randomly sample columns at each split
    "learning_rate": 1,  # Learning rate
    "max_depth": 5,  # Maximum depth of the trees
    "num_parallel_tree": 100,  # Number of trees to grow
    "objective": "binary:logistic",  # Objective function
    "subsample": 0.8,  # Randomly sample rows at each step
    "tree_method": "hist"  # Tree construction algorithm
}

bst = xgb.train(params, xg_data, num_boost_round=1)
```

```python
### Predcitions
xg_pred_data = DMatrix(X_test)
probabilities = bst.predict(xg_pred_data)
predictions = [True if prob >= 0.5 else False for prob in probabilities]

output = pd.DataFrame({"PassengerId" : test_data["PassengerId"], "Transported" : predictions})
output.to_csv("submission_sp_titanic.csv", index = False)
```
